#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingresos_tiendas = pd.DataFrame({ 'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
                                 'Ingresos Totales': [tienda['Precio'].sum(),tienda2['Precio'].sum(),
                                                      tienda3['Precio'].sum(),tienda4['Precio'].sum()]})
pd.options.display.float_format = '{:,.0f}'.format
ingresos_tiendas
ingresos_tiendas
```

```python
ingresos_tiendas = ingresos_tiendas.sort_values('Ingresos Totales', ascending=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))


plt.bar(ingresos_tiendas['Tienda'], ingresos_tiendas['Ingresos Totales'], color='#6A5ACD', alpha=0.7)


plt.plot(
    ingresos_tiendas['Tienda'],
    ingresos_tiendas['Ingresos Totales'],
    marker='o',
    color='black',
    linewidth=2
)

plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales')
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()
```

#5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
envio_t1 = tienda["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()
```

```python
envios_promedio = pd.DataFrame({"Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
                                "Envío Promedio": [round(envio_t1, 2),round(envio_t2, 2),round(envio_t3, 2),round(envio_t4, 2)]})

envios_promedio
```

```python
plt.figure(figsize=(10,6))
plt.bar(envios_promedio['Tienda'], envios_promedio['Envío Promedio'])
plt.title('Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Valor Promedio del Envío')
plt.xticks(rotation=0)
plt.show()
```

#6. Informe Final

El objetivo de este análisis es ayudar al Sr. Juan a identificar cuál de las cuatro tiendas de Alura Store presenta el rendimiento más bajo y, por lo tanto, sería la mejor candidata para vender. Para esto, se evaluaron los principales indicadores de desempeño de cada tienda, incluyendo: ingresos totales, categorías más vendidas, calificaciones promedio, productos más y menos vendidos y el costo promedio de envío.
A partir de estos datos, se busca identificar patrones, debilidades y oportunidades de mejora para tomar una decisión fundamentada.

1. Ingresos Totales por Tienda

Tienda 3: 1.098.020.000

La tienda con menor desempeño en ingresos es Tienda 4, mostrando una diferencia notable respecto a las demás.

2. Categorías más vendidas

En todas las tiendas se observaron diferencias importantes en el número de ventas por categoría. Cada tienda presenta una categoría dominante, pero la Tienda 4 muestra menor volumen general comparada con las demás, lo que evidencia menor rotación de productos.

3. Valoración promedio por tienda

El promedio de calificaciones fue:

Tienda 1: 4

Tienda 2: 4

Tienda 3: 4

Tienda 4: 4

Todas las tiendas presentan el mismo nivel de satisfacción por parte de los clientes, por lo que este indicador no influye directamente en la decisión.

4. Productos más y menos vendidos

En cada tienda se identificaron productos con mayor y menor número de ventas. Nuevamente, la Tienda 4 muestra menor variedad y menor volumen comparada con las demás, lo que refuerza la tendencia de bajo desempeño general.

5. Costo de envío promedio

El costo promedio de envío resultó muy similar entre las tiendas; sin embargo:

Tienda 4 reporta uno de los costos de envío más altos, lo cual puede reducir la competitividad y afectar las decisiones de compra.

6. Conclusión y Recomendación Final

Con base en todos los datos analizados, la tienda que El Sr. Juan debería vender es la Tienda 4.

Esta decisión se fundamenta en los siguientes puntos:

- Es la tienda con menores ingresos totales, mostrando un rendimiento inferior frente a las demás.
- Presenta menor volumen de ventas por categoría, lo que sugiere baja rotación y menor actividad comercial.
- Tiene la menor variedad y cantidad de productos más vendidos, evidenciando poco movimiento.

En conjunto, estos factores indican que la Tienda 4 es la menos eficiente y representa la mejor opción para vender, ya que su rendimiento actual es considerablemente inferior al del resto.
